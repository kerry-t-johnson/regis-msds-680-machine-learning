1. In a study of pleas and prison sentences, it is reported that
   42% of the subjects were sent to prison.
   Among those sent to prison, 38% plead guilty.
   Among those not sent to prison, 50% plead guilty.    

$$
\begin{aligned}
P(prison)                       = 0.42  \\
P(plea_{guilty} | prison)       = 0.38  \\
P(plea_{guilty} | \neg prison)  = 0.50  \\
\end{aligned}
$$


   a) If a subject is randomly selected, what is the probability of getting a
      person who was not sent to prison?    

      With no additional evidence to adjust our hypothesis, we calculate the
      probability based on P(prison):

$$
\begin{aligned}
      P(\neg prison)    &= 1 - P(prison)  \\
                        &= 1 - 0.42       \\
                        &= \mathbf{0.58}  \\
\end{aligned}
$$



   b) If a subject is randomly selected, and it is known that the subject
      entered a guilty plea, what is the probability that this subject was not
      sent to prison?    
   
      In this case, we have additional evidence (the subject entered a guilty
      plea) to adjust our hypothesis:

$$
\begin{aligned}
      P(\neg prison | plea_{guilty})      &= \frac{P(plea_{guilty}|\neg prison)P(\neg prison)}
                                                  {P(plea_{guilty}|\neg prison)P(\neg prison) + P(plea_{guilty}|prison)P(prison)} \\
                                          &= \frac{0.50 \times 0.58}
                                                  {\left(0.50 \times 0.58\right) + \left(0.38 \times 0.42\right)} \\
                                          &= \frac{0.29}{0.29 + 0.1596}       \\
                                          &= \mathbf{0.65}
\end{aligned}
$$



   c) If a subject is randomly selected, what is the probability of getting
      someone who was sent to prison?

      Similar to 1.a, we have no additional evidence to adjust our hypothesis.
      Therefore, this is a probability given in the initial problem description:
$$
   P(prison) = \mathbf{0.42}
$$



   d) If a subject is randomly selected, and it is known that the subject
      entered a guilty plea, what is the probability that this person was sent
      to prison?    

$$
\begin{aligned}     
      P(prison | plea_{guilty})     &= \frac{P(plea_{guilty}|prison)P(prison)}
                                            {P(plea_{guilty}|prison)P(prison) + P(plea_{guilty}|\neg prison)P(\neg prison)} \\
                                    &= \frac{0.38 \times 0.42}
                                            {\left(0.38 \times 0.42\right) + \left(0.50 \times 0.58\right)}     \\
                                    &= \frac{0.1596}{0.1596 + 0.29}     \\
                                    &= \mathbf{0.35}
\end{aligned}
$$


2. Given the following table:

|Customer ID| Gender  |Car Type|Shirt Size |Class|
|:---------:|:-------:|:------:|:---------:|:---:|
|1          |M        |Family  |Small      |CO   |
|2          |M        |Sports  |Medium     |CO   |
|3          |M        |Sports  |Medium     |CO   |
|4          |M        |Sports  |Large      |CO   |
|5          |M        |Sports  |Extra Large|CO   |
|6          |M        |Sports  |Extra Large|CO   |
|7          |F        |Sports  |Small      |CO   |
|8          |F        |Sports  |Small      |CO   |
|9          |F        |Sports  |Medium     |CO   |
|10         |F        |Luxury  |Large      |CO   |
|11         |M        |Family  |Large      |C1   |
|12         |M        |Family  |Extra Large|C1   |
|13         |M        |Family  |Medium     |C1   |
|14         |M        |Luxury  |Extra Large|C1   |
|15         |F        |Luxury  |Small      |C1   |
|16         |F        |Luxury  |Small      |C1   |
|17         |F        |Luxury  |Medium     |C1   |
|18         |F        |Luxury  |Medium     |C1   |
|19         |F        |Luxury  |Medium     |C1   |
|20         |F        |Luxury  |Large      |C1   |

a) What is the value of each of the following probabilities?<br>
$$
\begin{aligned}
   P(Gender=M | Class=C0)           &= \frac{6}{10}   \\
                                    &= \mathbf{0.6}   \\
                                                      \\
   P(Gender=F | Class=C1)           &= \frac{6}{10}   \\
                                    &= \mathbf{0.6}   \\
                                                      \\
   P(Car Type=Family | Class=C0)    &= \frac{1}{10}   \\
                                    &= \mathbf{0.1}   \\
                                                      \\
   P(Car Type=Family | Class=C1)    &= \frac{3}{10}   \\
                                    &= \mathbf{0.3}   \\
                                                      \\
   P(Shirt Size=Medium | Class=C0)  &= \frac{3}{10}   \\
                                    &= \mathbf{0.3}   \\
                                                      \\
   P(Shirt Size=Medium | Class=C1)  &= \frac{4}{10}   \\
                                    &= \mathbf{0.4}   \\
\end{aligned}
$$

b) Use Naive Bayes Classifier to find the class of P(Gender=F | Car Type=Family| Shirt Size=Medium)    
   $X = \left(Gender=F, Car Type=Family, Shirt Size=Medium\right)$

$$                                                      
\begin{aligned}
P(X|Class=C0)  &= P(Gender=F|Class=C0)                   \\ 
                  &\times P(Car Type=Family|Class=C0)    \\
                  &\times P(Shirt Size=Medium|Class=C0)  \\
               &= (1 - 0.6) \times 0.1 \times 0.3        \\
               &= 0.4 \times 0.1 \times 0.3              \\
               &= \mathbf{0.012}                         \\
                                                         \\
P(X|Class=C1)  &= P(Gender=F|Class=C1)                   \\
               &\times P(Car Type=Family|Class=C1)       \\
               &\times P(Shirt Size=Medium|Class=C1)     \\
               &= 0.6 \times 0.3 \times 0.4              \\
               &= \mathbf{0.072}                         \\
\end{aligned}
$$

Since $P(C0) = P(C1) = \frac{10}{20}$

Then...
$$
\begin{aligned}
P(X|Class=C1) \times P(C1) &\gt P(X|Class=C0) \times P(C0)  \\
P(X|Class=C1) &\gt P(X|Class=C0)
\end{aligned}
$$

Therefore: $\rightarrow Class = \mathbf{C1}$
